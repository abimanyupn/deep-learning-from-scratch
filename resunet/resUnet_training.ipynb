```python
# =========================
# 1. MOUNT GOOGLE DRIVE
# =========================

from google.colab import drive
drive.mount('/content/drive')

# =========================
# 2. IMPORT
# =========================

import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# =========================
# 3. PATH DATASET
# =========================

BASE = "/content/drive/MyDrive"

image_dir = os.path.join(
    BASE,
    "augmentasi 224 x 224",
    "stare-images-224x224 (augmented)"
)

mask_dir = os.path.join(
    BASE,
    "augmentasi 224 x 224",
    "labels-ah-224x224 (augmented)"
)

print("📂 IMAGE DIR:")
print(image_dir)

print("\n📂 MASK DIR:")
print(mask_dir)

# =========================
# 4. CEK FILE
# =========================

image_files = sorted(os.listdir(image_dir))
mask_files = sorted(os.listdir(mask_dir))

print("\n📊 TOTAL IMAGE :", len(image_files))
print("📊 TOTAL MASK  :", len(mask_files))

print("\n📄 SAMPLE IMAGE:")
print(image_files[:5])

print("\n📄 SAMPLE MASK:")
print(mask_files[:5])

# =========================
# 5. IMAGE SIZE
# =========================

IMG_SIZE = 224

# =========================
# 6. LOAD DATASET
# =========================

X = []
Y = []

print("\n📂 Loading dataset...")

for img_name, mask_name in zip(image_files, mask_files):

    # =========================
    # IMAGE
    # =========================

    img_path = os.path.join(image_dir, img_name)

    image = load_img(
        img_path,
        target_size=(IMG_SIZE, IMG_SIZE),
        color_mode="rgb"
    )

    image = img_to_array(image)
    image = image / 255.0

    # =========================
    # MASK
    # =========================

    mask_path = os.path.join(mask_dir, mask_name)

    mask = load_img(
        mask_path,
        target_size=(IMG_SIZE, IMG_SIZE),
        color_mode="grayscale"
    )

    mask = img_to_array(mask)
    mask = mask / 255.0

    X.append(image)
    Y.append(mask)

# convert ke numpy
X = np.array(X)
Y = np.array(Y)

print("\n✅ IMAGE SHAPE :", X.shape)
print("✅ MASK SHAPE  :", Y.shape)

# =========================
# 7. TRAIN VALID SPLIT
# =========================

split = int(len(X) * 0.8)

X_train = X[:split]
Y_train = Y[:split]

X_val = X[split:]
Y_val = Y[split:]

print("\n📊 TRAIN DATA :", len(X_train))
print("📊 VALID DATA :", len(X_val))

# =========================
# 8. RESNET50 ENCODER
# =========================

base_model = ResNet50(
    weights="imagenet",
    include_top=False,
    input_shape=(224, 224, 3)
)

# =========================
# 9. SKIP CONNECTION
# =========================

s1 = base_model.get_layer("input_layer").output
s2 = base_model.get_layer("conv1_relu").output
s3 = base_model.get_layer("conv2_block3_out").output
s4 = base_model.get_layer("conv3_block4_out").output

# bottleneck
b1 = base_model.get_layer("conv4_block6_out").output

# =========================
# 10. DECODER BLOCK
# =========================

def decoder_block(input_tensor, skip_tensor, num_filters):

    x = Conv2DTranspose(
        num_filters,
        (2, 2),
        strides=2,
        padding="same"
    )(input_tensor)

    x = Concatenate()([x, skip_tensor])

    x = Conv2D(
        num_filters,
        3,
        padding="same",
        activation="relu"
    )(x)

    x = Conv2D(
        num_filters,
        3,
        padding="same",
        activation="relu"
    )(x)

    return x

# =========================
# 11. DECODER
# =========================

d1 = decoder_block(b1, s4, 512)
d2 = decoder_block(d1, s3, 256)
d3 = decoder_block(d2, s2, 128)
d4 = decoder_block(d3, s1, 64)

# =========================
# 12. OUTPUT LAYER
# =========================

outputs = Conv2D(
    1,
    1,
    padding="same",
    activation="sigmoid"
)(d4)

# =========================
# 13. BUILD MODEL
# =========================

model = Model(
    inputs=base_model.input,
    outputs=outputs
)

# =========================
# 14. COMPILE MODEL
# =========================

model.compile(
    optimizer=Adam(1e-4),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

# =========================
# 15. MODEL SUMMARY
# =========================

model.summary()

# =========================
# 16. TRAINING
# =========================

history = model.fit(
    X_train,
    Y_train,
    validation_data=(X_val, Y_val),
    batch_size=4,
    epochs=10
)

# =========================
# 17. SAVE MODEL
# =========================

save_path = os.path.join(
    BASE,
    "resunet_retina_model.h5"
)

model.save(save_path)

print("\n🎉 MODEL SAVED:")
print(save_path)

# =========================
# 18. PREDICTION
# =========================

pred = model.predict(X_val[:1])

# =========================
# 19. VISUALIZATION
# =========================

plt.figure(figsize=(12,4))

# original retina
plt.subplot(1,3,1)
plt.imshow(X_val[0])
plt.title("Original Retina")

# ground truth
plt.subplot(1,3,2)
plt.imshow(Y_val[0].squeeze(), cmap="gray")
plt.title("Ground Truth")

# predicted mask
plt.subplot(1,3,3)
plt.imshow(pred[0].squeeze(), cmap="gray")
plt.title("Predicted Mask")

plt.show()
```
