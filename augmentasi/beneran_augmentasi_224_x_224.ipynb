```python
# 🔹 1. (opsional) mount drive — HAPUS kalau bukan di Colab
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# 🔹 2. import
import os
import gzip
from PIL import Image, ImageEnhance

# 🔹 3. setup path
BASE = "/content/drive/MyDrive"

# =========================
# CEK ISI FOLDER DATASET
# =========================

dataset_path = os.path.join(
    BASE,
    "dataset 224 x 224"
)

print("📂 ISI DATASET:")
print(os.listdir(dataset_path))

# =========================
# INPUT FOLDER
# =========================

input_folder = os.path.join(
    BASE,
    "dataset 224 x 224",
    "stare-images-224x224"
)

label_folder = os.path.join(
    BASE,
    "dataset 224 x 224",
    "labels-ah-224x224"
)

# =========================
# OUTPUT FOLDER
# =========================

output_base = os.path.join(
    BASE,
    "augmentasi 224 x 224"
)

# output image
output_folder = os.path.join(
    output_base,
    "stare-images-224x224 (augmented)"
)

# output label
output_label_folder = os.path.join(
    output_base,
    "labels-ah-224x224 (augmented)"
)

# bikin folder otomatis
os.makedirs(output_folder, exist_ok=True)
os.makedirs(output_label_folder, exist_ok=True)

# =========================
# CEK PATH
# =========================

print("\n📂 IMAGE PATH:", input_folder)
print("✅ IMAGE EXISTS:", os.path.exists(input_folder))

print("\n📂 LABEL PATH:", label_folder)
print("✅ LABEL EXISTS:", os.path.exists(label_folder))

if not os.path.exists(input_folder):
    print("\n❌ path image salah")
    raise SystemExit

if not os.path.exists(label_folder):
    print("\n❌ path label salah")
    raise SystemExit

print("\n📄 SAMPLE IMAGE FILE:")
print(os.listdir(input_folder)[:5])

print("\n📄 SAMPLE LABEL FILE:")
print(os.listdir(label_folder)[:5])

# =========================
# AUGMENT FUNCTION
# =========================

def augment_pencil_cells(img):

    angles = [45, 90, 135, 180, 225, 270, 315]
    brightness_factors = [0.8, 1.2]

    results = []

    # original
    results.append(("original", img))

    # flip horizontal
    results.append((
        "flip_horizontal",
        img.transpose(Image.FLIP_LEFT_RIGHT)
    ))

    # flip vertical
    results.append((
        "flip_vertical",
        img.transpose(Image.FLIP_TOP_BOTTOM)
    ))

    # rotate
    for angle in angles:

        rot_image = img.rotate(
            angle,
            expand=False,
            fillcolor=0
        )

        results.append((
            f"rotate_{angle}",
            rot_image
        ))

    # brightness
    temp_results = results.copy()

    for name, image in temp_results:

        enhancer = ImageEnhance.Brightness(image)

        for factor in brightness_factors:

            factor_name = str(factor).replace(".", "")

            bright_image = enhancer.enhance(factor)

            results.append((
                f"{name}_brightness_{factor_name}",
                bright_image
            ))

    return results

# =========================
# IMAGE LOADER
# =========================

def load_image(path):

    if path.endswith(".gz"):

        with gzip.open(path, 'rb') as f:

            img = Image.open(f)
            img = img.convert("L")
            img.load()

    else:

        img = Image.open(path)
        img = img.convert("L")

    return img

# =========================
# PROCESS DATASET
# =========================

def process_dataset(input_dir, output_dir):

    files = os.listdir(input_dir)

    print(f"\n📊 TOTAL FILE: {len(files)}")
    print(f"📂 PROCESSING: {input_dir}")

    count = 0

    valid_extensions = (
        ".ppm",
        ".ppm.gz",
        ".png"
    )

    for filename in files:

        if filename.endswith(valid_extensions):

            path = os.path.join(input_dir, filename)

            try:

                img = load_image(path)

                augmented_images = augment_pencil_cells(img)

                base_name = (
                    filename
                    .replace(".ppm.gz", "")
                    .replace(".ppm", "")
                    .replace(".png", "")
                )

                for name, aug_img in augmented_images:

                    save_path = os.path.join(
                        output_dir,
                        f"{base_name}__{name}.png"
                    )

                    aug_img.save(save_path)

                count += 1

                print(f"✅ {count} processed: {filename}")

            except Exception as e:

                print(f"⚠️ error di {filename}: {e}")

    print(f"\n🎉 DONE! total processed: {count}")

# =========================
# RUN
# =========================

# augment image
process_dataset(
    input_folder,
    output_folder
)

# augment label
process_dataset(
    label_folder,
    output_label_folder
)
```
